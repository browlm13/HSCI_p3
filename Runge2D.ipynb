```python
%pylab inline
Runge = loadtxt('Runge.txt')

a = loadtxt('avals.txt')
b = loadtxt('bvals.txt')

p6_uni = loadtxt('p6_uni.txt')
p24_uni = loadtxt('p24_uni.txt')

p6_Cheb = loadtxt('p6_Cheb.txt')
p24_Cheb = loadtxt('p24_Cheb.txt')
```

# uniformly spaced nodes

## Runge function

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D

fig = figure()

ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('x')
ax.set_ylabel('y')
title('Runge with uniformly spaced nodes')

X, Y = meshgrid(b, a)
surf = ax.plot_surface(X, Y, Runge, rstride=1, cstride=1, linewidth=0, cmap=cm.jet)
```

## n = 6

```python
fig = figure()

ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('x')
ax.set_ylabel('y')
title('$p_6$ with uniformly spaced nodes')

X, Y = meshgrid(b, a)
surf = ax.plot_surface(X, Y, p6_uni, rstride=1, cstride=1, linewidth=0, cmap=cm.jet)
```

```python
#calculate error
error_6_uni=numpy.ndarray((201,101))
for i in range(0,201):
    for j in range(0,101):
      error_6_uni[i][j] = abs(Runge[i][j]- p6_uni[i][j])
    
fig = figure()

ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('x')
ax.set_ylabel('y')
title('Error of $p_6(x)$ with uniformly spaced nodes')

X, Y = meshgrid(b, a)
surf = ax.plot_surface(X, Y, error_6_uni, rstride=1, cstride=1, linewidth=0, cmap=cm.jet)
```
